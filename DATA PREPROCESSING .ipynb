## **Handle Categorical Values**

### **1- One Hot Encoding (2 category)**

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df['Sex'].unique()
```

```python
#one hot encoding
sex_converted = pd.get_dummies(df['Sex'], drop_first= False, dtype = 'int')
```

```python
sex_converted
```

```python
df.head()
```

```python
# add sex converted to df
df = pd.concat([df, sex_converted], axis = 1)
df.head()
```

```python
df.drop(['Sex'], axis = 1, inplace = True)
```

```python
df.head()
```

### **2- One Hot Encoding with Multiple Categories (more than 2 categories)**

```python
mercendez_df = pd.read_csv("mercendez.csv")
mercendez_df.head()
```

```python
#check for unique labels in each column
for col in mercendez_df.columns:
    print(len(mercendez_df[col].unique()))
```

```python
mercendez_df.X1.value_counts()
```

```python
list_categories = mercendez_df.X1.value_counts().sort_values(ascending = False)
list_categories
```

```python
for category in list_categories:
    mercendez_df[category] = np.where(mercendez_df['X1'] == category, 1,0)
```

```python
mercendez_df.head()
```

```python
mercendez_df.drop('X1', axis = 1, inplace = True)
mercendez_df.head()
```

### **Ordinal Number Encoding**

```python
data = {'Temperature' : ['Hot', 'Cold', 'Very Hot', 'Cold', 'Warm', 'Hot', 'Warm']}
```

```python
data = pd.DataFrame(data, columns = ['Temperature'])
data
```

```python
data['Temperature'].unique()
```

```python
Temperature_values = {'Cold': 1, 'Warm':2, 'Hot':3, 'Very Hot':4}
```

```python
Temperature_values
```

```python
#map between temperature and Temperature_values
data['Temperature_Encoding'] = data.Temperature.map(Temperature_values)
```

```python
data
```

```python
data.drop('Temperature', axis = 1, inplace = True)
data
```

### **4- Count or Frequency Encoding**

```python
adult_df = pd.read_csv("adult.data", header = None, index_col = None)
```

```python
adult_df.head()
```

```python
columns = [1,3,5,6,7,8,9,13]
```

```python
adult_df = adult_df[columns]
adult_df.head()
```

```python
adult_df.columns = ['Employment', 'Degree', 'Status', 'Designation', 'Family_Job', 
                    'Race', 'Gengder', 'Country']
```

```python
for col in adult_df.columns[:]:
    print(col, ":", len(adult_df[col].unique()), "category")
```

```python
#map each country with it's count
country_map = adult_df['Country'].value_counts().to_dict()
country_map
```

```python
adult_df['Country'] = adult_df['Country'].map(country_map)
adult_df.head()
```

### **5- Target Guided Ordinal Encoding**

```python
df = pd.read_csv('titanic.csv', usecols = ['Survived', 'Cabin'])
df.head()
```

```python
df.info()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].fillna("Missing", inplace = True)
df.head()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head()
```

```python
ordinal_index = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_index
```

```python
ordinal_map = {k:v for v,k in enumerate(ordinal_index,0)}
ordinal_map
```

```python
df['Cabin_Encoded'] = df['Cabin'].map(ordinal_map)
df.head()
```

### **6- Mean Ordinal Encoding**

```python
df.drop('Cabin_Encoded', axis = 1, inplace = True)
```

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_ordinal'] = df['Cabin'].map(mean_ordinal)
df
```

### **7- Probability Patio Encoding**

```python
df.drop('mean_ordinal', axis = 1, inplace = True)
df.head()
```

```python
prob = df.groupby(['Cabin'])['Survived'].mean()
prob
```

```python
prob_df = pd.DataFrame(prob)
prob_df
```

```python
prob_df['Died'] = 1-prob_df['Survived']
prob_df
```

```python
prob_df['Probability Ratio'] = prob_df['Survived'] / prob_df['Died']
prob_df
```
